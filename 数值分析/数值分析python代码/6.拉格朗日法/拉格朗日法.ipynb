```python
import matplotlib.pyplot as plt
import numpy as np
#忽略警告
import warnings
warnings.filterwarnings("ignore")

def func(x):
    return 0.5*x**5+x**4+x**3-5*x**2-4*x+1
    
def Lagrangian(x):
    global xList
    length = len(xList)
    number = 0
    for i in range(length):
        temp = 1
        for j in range(length):
            if i == j:
                continue
            numb = (x - xList[j])/(xList[i] - xList[j])
            temp *= numb
        temp *= func(xList[i])
        number += temp
    return number
    
global xList
xList = list()

n = int(input('请输入点的个数:'))
for i in range(n):
    num = np.random.uniform(-2.5, 3)
    xList.append(num)
    
#第一行将显示中文字符的默认字体设置为"SimHei"，这是一种常用的中文字体。当生成带有中文标签或注释的图表时，这非常有用。
#第二行确保Matplotlib不会将负号显示为方框或其他Unicode字符。通过将axes.unicode_minus设置为False，Matplotlib使用标准的连字符字符来表示负号。
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams['axes.unicode_minus'] = False


line = np.arange(-3,3.5,0.1)

# plt.figure(figsize=(8,18))：这行代码创建了一个图形窗口，并设置其尺寸为宽度8英寸、高度18英寸。通过指定figsize参数，可以调整图形窗口的大小。
plt.figure(figsize=(8,18))

# plt.subplot(3,1,1)：这行代码选择第一个子图进行操作。其中，3表示总共有3行子图，1表示总共有1列子图，1表示当前操作的是第1个子图。通过使用plt.subplot()函数，可以在图形窗口中创建多个子图，并选择特定的子图进行绘制或操作。
plt.subplot(3,1,1)
plt.plot(line, func(line), color = 'g',label = '原函数')
plt.scatter(np.array(xList),func(np.array(xList)),color = 'r')
plt.legend(['原函数'])
plt.grid()
plt.title('原函数和点')


plt.subplot(3,1,2)
plt.plot(line, Lagrangian(line), color = 'b',label = '拟合函数')
plt.scatter(np.array(xList),func(np.array(xList)),color = 'r')
plt.legend(['拟合函数'])
plt.grid()
plt.title('拟合函数')


plt.subplot(3,1,3)
plt.plot(line, Lagrangian(line), color = 'b',label = '拟合函数')
plt.plot(line, func(line), color = 'g',label = '原函数')
plt.scatter(np.array(xList),func(np.array(xList)),color = 'r')
plt.grid()
plt.legend(['拟合函数','原函数'])
plt.title('两曲线重合度')
'''
最好效果为6个点！
'''
```
